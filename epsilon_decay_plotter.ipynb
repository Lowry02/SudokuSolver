```python
import numpy as np
import matplotlib.pyplot as plt

def plot_epsilon_decay(a, n_max=500):
  n = np.arange(0, n_max + 1)
  y = 0.99 * a ** n
  plt.figure(figsize=(20,10))
  plt.plot(n, y)
  plt.xlabel('n')
  plt.ylabel('0.99 * a ** n')
  plt.title(f'Plot of 0.99 * a ** n (a={a})')
  plt.grid(True)
  plt.show()

# Example usage:
# plot_function(a=1.05)
```

```python
plot_epsilon_decay(0.96)
```
